```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('Students.csv')
data.head()
```

```python
data['MathScore'].fillna(data['MathScore'].mean(), inplace = True)
data['EnglishScore'].fillna(data['EnglishScore'].mean(), inplace = True)
data.head()
```

```python
data['TotalScore'] = data['MathScore'] + data['EnglishScore']
data
```

```python
scaler = StandardScaler()
data[['MathScore', 'EnglishScore', 'TotalScore']] = scaler.fit_transform(data[['MathScore', 'EnglishScore', 'TotalScore']])
data
```

```python
le = LabelEncoder()

data['Gender'] = le.fit_transform(data['Gender'])
data
```
