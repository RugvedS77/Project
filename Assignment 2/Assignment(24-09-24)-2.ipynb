```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Housing.csv')
```

```python
sample_data = data.iloc[::30, :]

x = sample_data['area']
y = sample_data['price']

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Price vs Area (Sampled Data)')
plt.show()
```

```python
furnished_avg = data[data['furnishingstatus'] == 'furnished']['price'].mean()
unfurnished_avg = data[data['furnishingstatus'] == 'unfurnished']['price'].mean()
x=["furnished","unfurnished"]
y = [furnished_avg,unfurnished_avg]

plt.bar(x, y, color=['red', 'yellow'])

plt.xlabel('Furnishing Status')
plt.ylabel('Average Price')
plt.title('Average Price by Furnishing Status')

plt.show()
```

```python
data.boxplot(column='price', by='stories')
plt.xlabel('Stories')
plt.ylabel('Price ($)')
plt.title('Price Distribution by Number of Stories')
plt.show()
```

```python
mainroad_counts = data['mainroad'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(mainroad_counts, labels=mainroad_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title('Proportion of Houses on Main Road vs. Not')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['area'], bins=20, color='purple', edgecolor='black')
plt.title('Distribution of House Areas')
plt.xlabel('Area (sq ft)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
